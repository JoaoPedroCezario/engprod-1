#Função de várias variáveis
Introdução

```python
from sympy import * #importando a biblioteca sympy
from sympy.plotting import * #plot de gráficos
init_printing(use_unicode=True)
x, y, z = symbols('x y z')
```

```python
plot3d(1+x**2+y**2) #exemplo de gráfico, vamos ver se deu tudo ok
```

Problema inicial

```python
demanda_A = 1300 - 50*x + 20*y
demanda_A 
```

```python
demanda_B = 1700 + 12*x - 20*y
demanda_B
```

```python
receita_A = x*demanda_A
```

```python
receita_A
```

```python
simplify(receita_A)
```

```python
receita_B = y*demanda_B
```

```python
receita_B
```

```python
receita_total = receita_A + receita_B
receita_total
```

```python
simplify(receita_total)
```
