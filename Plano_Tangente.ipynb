#Plano Tangente

```python
from sympy import *
from sympy.plotting import *
init_printing(use_unicode=True)
x, y, z = symbols('x y z')
```

```python
# Valores de entrada

P = (0, 0, 3)
f = (2*x+3)/(4*y+1)

# Encontrando a equção do plano tangente de f em P:

fx = diff(f, x).subs({x: P[0], y: P[1]})
fy = diff(f, y).subs({x: P[0], y: P[1]})

plano_tangente = P[2]+fx*(x-P[0])+fy*(y-P[1])

plano_tangente
```

```python
grafico = plot3d(f, (x, -4, 4), (y,-4, 4), show = False)
plano = plot3d(plano_tangente, (x, -4, 4), (y,-4, 4), show = False)

grafico.append(plano[0])

grafico.show()
```
